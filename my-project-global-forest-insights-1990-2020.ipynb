Top 10 Countries with Highest Forest Area in 2020

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data
forest_df = pd.read_csv('/kaggle/input/global-forest-insights-1990-2020-fao/Forest_Area.csv')

# Preprocessing and sorting
forest_df['Forest Area, 2020'] = pd.to_numeric(forest_df['Forest Area, 2020'],errors='coerce')
forest_df = forest_df.nlargest (10,'Forest Area, 2020')

# Prepare data for pie chart
labels = forest_df['Country and Area']
sizes = forest_df['Forest Area, 2020']

# Plotting pie chart
plt.figure(figsize=(10,7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Adding title
plt.title('Top 10 Countries with Highest Forest Area in 2020')

plt.show()
```

The percentages displayed on each slice represent the proportion of the total forest area occupied by that specific country among the top 10 countries.
